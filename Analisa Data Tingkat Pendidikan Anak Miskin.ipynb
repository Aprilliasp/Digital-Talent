## Tingkat Pendidikan Anak Miskin
## Deskripsi
Dataset ini berisi data mengenai jumlah anak dengan status kesejahteraan 30% yang bersekolah berdasarkan tingkat pendidikan dan jumlah individu:

    1. Kode_Provinsi : Kode Provinsi
    2. Nama_Provinsi : Nama Provinsi
    3. Tingkat_Pendidikan : Tingkat Pendidikan (SD/MI/SMP/MT/SMA/MA/Perguruan Tinggi)
    4. Jenis_kelamin : Jenis Kelamin
    5. Jumlah_individu : Jumlah dari Anak dengan Tingkat Pendidikan dan Jenis Kelamin tersebut





```python
import pandas as pd
```

```python
termiskin = pd.read_csv('termiskin.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
termiskin
```

## Statistik

```python
print('Jumlah Tingkat Pendidikan Anak Miskin yang didata adalah',termiskin.jumlah_individu.sum())
```

```python
termiskin.kode_provinsi.count()
```

```python
a = termiskin.jumlah_individu.min()
b = termiskin.jumlah_individu.max()

print('Jumlah Minimal Tingkat Pendidikan Anak Miskin',a)
print('Jumlah Maksimal Tingkat Pendidikan Anak Miskin',b)
```

```python
me = termiskin.mean()
md = termiskin.median()

print(me)
```

```python
print(md)
```

## Visualisasi

```python
ax = termiskin.groupby('tingkat_pendidikan').mean()['jumlah_individu']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Tingkat Pendidikan", legend=True, fontsize=12)
ax.set_xlabel("tingkat_pendidikan", fontsize=12)
ax.set_ylabel("jumlah_individu", fontsize=12)
```

```python
ax = termiskin.groupby('nama_provinsi').mean()['jumlah_individu']
ax = ax.plot(kind='line', title ="Rata-Rata Berdasarkan ", legend=True, fontsize=12)
ax.set_xlabel("nama_provinsi", fontsize=12)
ax.set_ylabel("jumlah_individu", fontsize=12)
```

```python
ax = termiskin.groupby('nama_provinsi').mean()['jumlah_individu']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan ",figsize=(10,5), legend=True, fontsize=12)
ax.set_xlabel("nama_provinsi", fontsize=12)
ax.set_ylabel("jumlah_individu", fontsize=12)
```
